### Reclassifying target

```python
df['quality'] = df['quality'].replace([3,4,5],'low')
df['quality'] = df['quality'].replace([6],'medium')
df['quality'] = df['quality'].replace([7,8,9],'high')


df['quality'] = df['quality'].astype('category')
```

---

### Splitting and encoding features

```python
X = df.drop(columns = ['Wine','quality',])
y = df['quality']

le = LabelEncoder()
le.fit(y)

y = le.transform(y)
```

---

### Random under sampler

```python
rus = RandomUnderSampler(random_state=42)
X_res, y_res = rus.fit_resample(X, y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_res, y_res,random_state=42)
```

```python
rfc = RandomForestClassifier(
                            max_depth = 15,
                            max_features = 'sqrt', 
                            #max_leaf_nodes = 1000,
                            min_samples_split = 4,
                            n_estimators = 100
)
```

```python
rfc.fit(X_res, y_res)
```

```python
y_pred = rfc.predict(X_test)

cm = confusion_matrix(y_test, y_pred)

display = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=le.classes_)
display.plot()
plt.show()
```

```python
print(classification_report(y_test,y_pred))
```
